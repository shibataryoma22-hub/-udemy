with構文
ファイルの保存や読み込み

```python
#ファイル保存
greetings = "Good morning\nGood afternoon\nGood night"

with open("greetings.txt", "w") as f:
  f.write(greetings)
```

```python
#ファイルの読み込み
with open("greetings.txt", "r") as f:
  print(f.read())
```
