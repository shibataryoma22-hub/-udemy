python基礎

```python
a = [1, 2, 3, 4, 5] #リストの作成

b = a[2] #3番目の要素を取得
print(b)

a.append(6) #末尾に要素を追加
print(a)
a[2] = 7 #要素の入れ替え
print(a)

#タプル 要素の追加、削除、入れ替えはできない
a = (1, 2, 3, 4, 5)
b = a[2]
print(b)
```

```python
#要素数１のタプル
a = (3,)
```

```python
#辞書
a = {"Artificial":1, "Intelligence":2}
print(a["Artificial"]) #Artificialのキーを持つ値を取得

a["Artificial"] = 7 #要素の入れ替え
print(a["Artificial"])
a["ML"] = 3 #要素の追加
print(a)
```

```python
#セット
#重複した値の要素をもたない
#要素の追加、削除可能
a = [1, 1, 2, 3, 4, 4 ,5, 5, 5] #リストa
print(a)

b = set(a) #セットに変換
print(b)

b.add(6) #要素の追加
print(b)


```

内包表記
リストの要素を操作したうえで、新しいリストの作成するための記法
新たなリスト = [要素への処理 for 要素 in リスト]

```python
a = [1, 2, 3, 4, 5, 6]
b = [c*3+1 for c in a] #aの要素を３倍して１を足し新たなリストを作る
print(b)
```
