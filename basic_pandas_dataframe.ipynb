DataFrameの作成
DataFrameはラベル付きの二次元配列で、整数や少数、文字列などさまざまな型のデータを格納することができる

```python
import pandas as pd
import numpy as np
```

```python
a = pd.DataFrame([[80, 60, 70, True],
                  [90, 80, 70, True],
                  [70, 60 ,50, True],
                  [40, 60, 50, False],
                  [20, 30, 40, False],
                  [50, 20, 10, False]])
```

```python
print(a)
```

```python
a.index = ["1", "2", "3", "4", "5", "6"]
a.columns = ["Japanese", "Math", "English", "Result"]
print(a)
```

```python
a.shape #行列
```

```python
a.head() #最初の５行
```

```python
a.tail() #最後の５行
```

```python
a.describe() #基本的な統計量
```

```python
#DataFrameの操作
tr = a.loc["1", :] #１行取り出す
print(type(tr))
tr
```

```python
ma = a.loc[:, "English"]
print(type(ma))
ma
```

```python
#ilocによって範囲指定可能になる
r = a.iloc[1:4, :2] #1行目から３行目まで、０－１列
print(type(r))
r
```

```python
#locメソッドによる行の追加
a.loc["Shiro"] = pd.Series([70, 80, 70, True], index = ["Japanese", "Math", "English", "Result"], name = "Shiro")
```

```python
print(a)
```

```python
#列ラベルの指定で列を追加できる
a["Science"] = [80, 70, 60, 50, 60, 40, 80]
a
```

```python
a.sort_values(by="Math", ascending=False)
```
