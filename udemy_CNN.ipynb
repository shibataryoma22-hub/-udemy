```python
from torchvision.datasets import CIFAR10
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
```

```python
cifar10_data = CIFAR10(root = "./data",
                       train = True,
                       download = True,
                       transform = transforms.ToTensor())
cifar10_classes = np.array(["airplane", "automobile", "bird", "cat", "deer",
                            "dog", "frog", "horse", "ship", "truck"])

print("データの数:", len(cifar10_data))

n_image = 25 #表示する画像の数
cifar10_loader = DataLoader(cifar10_data,
                            batch_size = n_image,
                            shuffle = True)
dataiter = iter(cifar10_loader) #データローダから１枚ずつ画像を取り出せる
images, labels = dataiter.next() #最初のバッチを取り出す
plt.figure(figsize = (10, 10))
for i in range(n_image):
  plt.subplot(5, 5, i+1)
  plt.imshow(np.transpose(images[i], (1, 2, 0)))
  label = cifar10_classes[labels[i]]
  plt.title(label)
  plt.tick_params(labelbottom = False, labelleft = False, bottom = False, left = False)

plt.show()
```

```python
transform = transforms.Compose(transforms.RandomAffine([-30, 30], scale = (0.8, 1.2)),
                               transforms.ToTensor())


cifar10_data = CIFAR10(root = "./data",
                       train = True,
                       download = True,
                       transform = transforms.ToTensor())
cifar10_classes = np.array(["airplane", "automobile", "bird", "cat", "deer",
                            "dog", "frog", "horse", "ship", "truck"])

print("データの数:", len(cifar10_data))

n_image = 25 #表示する画像の数
cifar10_loader = DataLoader(cifar10_data,
                            batch_size = n_image,
                            shuffle = True)
dataiter = iter(cifar10_loader) #データローダから１枚ずつ画像を取り出せる
images, labels = dataiter.next() #最初のバッチを取り出す
plt.figure(figsize = (10, 10))
for i in range(n_image):
  plt.subplot(5, 5, i+1)
  plt.imshow(np.transpose(images[i], (1, 2, 0)))
  label = cifar10_classes[labels[i]]
  plt.title(label)
  plt.tick_params(labelbottom = False, labelleft = False, bottom = False, left = False)

plt.show()
```

```python
from torch._higher_order_ops.invoke_subgraph import trace_joint_graph
affine = transforms.RandomAffine([-15, 15], scale = (0.8, 1.2))
flip = transforms.RandomHorizontalFlip(p=0.5) #50%の確率で反転
normalize = transforms.Normalize((0.0, 0.0, 0.0), (1.0, 1.0, 1.0)) #平均値０、標準偏差１ RGB
to_tensor = transforms.ToTensor()

transform_train = transforms.Compose([affine, flip, normalize, to_tensor])
transform_test = transforms.Compose([to_tensor, normalize])
cifar10_train = CIFAR10("./dawta",
                        train = True,
                        download = True,
                        transform = transform_train)
cifar10_test = CIFAR10("./dawta",
                       train = False,
                       download = True,
                       transform = transform_test)
#DataLoaderの設定
batch_size = 64
train_loader = DataLoader(cifar10_train,
                          batch_size= batch_size,
                          shuffle = True)
test_loader = DataLoader(cifar10_test,
                         batch_size = len(cifar10_test),
                         shuffle = False)
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):
  def __init__(self):
    super(Net,self).__init__()
    self.conv1 = nn.Conv2d(3, 6, 5) #入力チャネル数，フィルタ数，フィルタサイズ
    self.pool = nn.MaxPool2d(2, 2) #プーリング層（領域のサイズ，ストライド）
    self.conv2 = nn.Conv2d(6, 16, 5)
    self.fc1 = nn.Linear(16*5*5, 256)
    self.dropout = nn.Dropout(p=0.5)
    self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.self.conv2(x))
      x = x.view(-1, 16*5*5) #
      x = F.relu(self.fc1(x))
      y = F.relu(self.fc2(x))
      return y
```

```python
net = Net()
net.cuda()
print(net)
```

```python
#学習A
from torch import optim
#交差エントロピー誤差
loss_fnc = nn.CrossEntropyLoss()
#最適化関数
optimizer = optim.Adam(net.parameters())
```

```python
#損失関数のloss
record_loss_train = []
record_loss_test = []

x_test, t_test = iter(test_loader).next()
x_test, t_test = x_test.cuda(), t_test.cuda()

for i (x, t), in enumerate(train_loader):
  x, t = x.cuda(), t.cuda()
  y = net(x)
  loss = loss_fnc(y, t)
  loss_train += loss.item()
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
loss_train /= i+1
record_loss_train.append(loss_train)

net.eval()
y_test = net(x_test)
loss_test = loss_fnc(y_test, t_test).item()
record_loss_test.append(loss_test)

if i%1 == 0:
  print("Epoch", i, "Loss_train:", loss_train, "Loss_test", loss_test)
```

```python
plt.plot(range(len(record_loss_train)), record_loss_train, label = "Train")
plt.plot(range(len(record_loss_test)), record_loss_test, label = "Test")
plt.legend()
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
correct = 0
total = 0

net.eval()
for i, (x, t) in enumerate(test_loader):
  x, t = x.cuda(), t.cuda()
  y = net(x)
  correct += (y.argmax(1), == t).sum().item()
  total += len(x)
print("正解率:", str(correct / total * 100) + "%")
```
