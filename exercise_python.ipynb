reshapeによる配列形状の操作

```python
import numpy as np
a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([[5, 4, 3], [2, 1, 0]])

a = a.reshape(2, 3)
print(a + b)
```

y = x^3 - 12x のグラフを描画

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5)
y = x ** 3 - 12 * x

plt.plot(x, y) #グラフの描画
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
