```python
#関数の作成
def add(a, b):
  c = a + b
  return c

print(add(3, 4))

def add(a, b=4):#第２引数にデフォルト値4を設定
   c = a + b
   return c
print(add(3))

def add(a, b, c):
  d = a + b + c
  return d

e = (1, 2, 3)
print(add(*e)) #*を使って引数を一気に渡せる
```

関数内で定義された変数がローカル変数、関数外で定義された関数がグローバル変数
ローカル変数→同じ関数内からのみ参照可能
グローバル変数→どこからでも可能

```python
a = 123 #グローバル変数
def showNum():
  b = 456 #ローカル変数
  print(a, b)

showNum()
```

```python
a = 123
def setLocal():
  a = 456 #aはローカル変数とみなされる
  print("Local:", a)

setLocal()
print("Global", a)
```

```python
a = 123
def setGlobal():
  global a #aをグローバル変数と明記することで関数内でグローバル変数の値を変更可能
  a = 456
  print("Global", a)
setGlobal()
print("Global", a)
```
