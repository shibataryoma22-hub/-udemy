pythonはオブジェクト指向のプログラミング言語
クラスとインスタンスがある
クラスは設計図
インスタンスは実体

クラスから複数のインスタンスを作成可能
クラスとインスタンスを総称してオブジェクトという

クラスで複数のメソッドをまとめることができる

```python
class Calc:
  def __init__(self, a): #コンストラクタ、初期設定
    self.a = a

  def add(self, b):
    print(self.a + b)

  def multiply(self, b):
    print(self.a * b)

```

```python
calc = Calc(3) #3はinitメソッドに渡す値
calc.add(4)
calc.multiply(4)
```

継承
既存のクラスを引き継いで新しいクラスを定義できる

```python
class CalcPlus(Calc): #Calcを継承
  def subtract(self, b):
    print(self.a - b)
  def divide(self, b):
    print(self.a / b)

```

```python
calc_plus = CalcPlus(3)
calc_plus.add(4)
calc_plus.multiply(4)
calc_plus.subtract(4)
calc_plus.divide(4)
```

__call__メソッド
インスタンス名からメソッドを呼び出し可能

```python
class Hello:
  def __init__(self, name):
    self.name = name
  def __call__(self):
    print("Hello" + self.name + "!")
h = Hello("AI")
h()

Hello("AI")() #h()と一緒
```
